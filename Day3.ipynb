```python
# Using .apply()
# .apply() goes into the list and gets every value
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

df["sum"] = df.apply(lambda row: row["A"] + row["B"], axis = 1)
print(df)
```

```python
# import pandas as pd

# Dataframe
data = {
    "fruit" : ["low", "med", "high", "low", "high", "unkown"]
}

# Making the actual DataFrame
df = pd.DataFrame(data)

# translating the words into dic numbers
category_map = {"low" : 0, "med" : 1, "high" : 2}

# using .apply()
df["enc"] = df["fruit"].apply(lambda x: category_map.get(x, -1))

# Showing the new DataFrame
print(df)
```

```python
# import pandas as pd

# Data
df = pd.DataFrame({
    "value": [10, 20, 30]
})

# mean and std
mean = df["value"].mean() # 20
std = df["value"].std() # 10

# Feature scaling
df["norm"] = df["value"].apply(lambda x: (x - mean) / std)
print(df)
```

```python
# import pandas as pd
df = pd.DataFrame({
    "text": ["Hello world", "Python is fun"]
})

for index, row in df.iterrows():
    print("Index:", index)
    print("Text:", row["text"])
```

```python
# Vectorized

# Data
df = pd.DataFrame({
    "value": [10, 20, 30],
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

# Vectorizing
df["Double"] = df["value"] * 2
df["add"] = df["A"] + df["B"]

# Printing
print(df["Double"])
print(df["add"])
```

```python
# Vectorized
# import pandas as pd

df = pd.DataFrame({
    "text" : ["Hello world", "Python is fun", "I love AI"]
})

# Using the fast way
df["text_length"] = df["text"].str.split().str.len()

# Printing the new DataFrame
print(df)
```
