```python
df = pd.read_csv("Data/day17_scaling.csv")
numcols = ["CRIM", "RM"]
minmax = MinMaxScaler()
standard = StandardScaler()
robust = RobustScaler()
minmax_vals = minmax.fit_transform(df[numcols])
standard_vals = standard.fit_transform(df[numcols])
robust_vals = robust.fit_transform(df[numcols])
print("MinMax mean :", minmax_vals.mean(axis = 0))
print("Standard mean / std :", standard_vals.mean(axis = 0), standard_vals.std(axis = 0))
print("Robust median approx :", pd.DataFrame(rob_vals).median().tolist())
```
