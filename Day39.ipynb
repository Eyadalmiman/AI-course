**Code**

```python
X = np.array([
[3.0, 4.0],
[1.0, 2.0],
[0.0, 5.0],
])

row_norms = np.linalg.norm(X, axis = 1, keepdims = True)
X_normalized = X / row_norms

print(row_norms)
print(X_normalized)
```
