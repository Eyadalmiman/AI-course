```python
# Assignment 1
# from typing import List, Dict, Tuple
# import pandas as pd
data1 = [10, 20, 30]
data2 = [40, 50, 60]

def calculate_stats(data:list[float])->dict[str, float]:
    mean = sum(data) / len(data)
    max_number = max(data)
    min_number = min(data)
    results = {"mean" : mean,
               "max" : max_number,
               "min" : min_number}
    return results

stats_a = calculate_stats(data1)
stats_b = calculate_stats(data2)

print("A:", stats_a)
print("B:", stats_b)
```
