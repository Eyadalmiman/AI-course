```python
# Activity 1
# from typing import List, Dict, Tuple
cat_a = [10, 20, 30]
cat_b = [40, 50, 60]

def calculate_stats(data:List[float])->Dict[str, float]:
    mean = sum(data) / len(data)
    max_number = max(data)
    min_number = min(data)
    results = {"mean" : mean,
               "max" : max_number,
               "min" : min_number}
    return results

stats_a = calculate_stats(cat_a)
stats_b = calculate_stats(cat_b)
print("A:", stats_a)
print("B:", stats_b)
```

```python
# Activity 2
# from typing import List, Dict, Tuple
# import math
data = [10, 20, 30]
def normalize_data(data:List[float], method:str=["minmax"])->List[float]:
    if not data:
        return []
    
    if method == "minmax":
        valuemin = min(data)
        valuemax = max(data)

        if valuemax == valuemin:
            return [0.0 for x in data]
        
        return [(x - valuemin) / ((valuemax - valuemin) for x in data)]
    
    elif method == "zscore":
        mean = sum(data) / len(data)
        std = sum((x - mean) ** 2 for x in data / len(data)) ** 0.5
        squareroot = math.sqrt(std)

        if squareroot == 0:
            return [0.0 for x in data]
        
        return [(x - mean) / squareroot for x in data]
    else:
        raise ValueError("Method must be 'minmax' or 'zscore'")

nor_stan = normalize_data(data, "minmax")
print(nor_stan)
```

```python
# Activity 3
# from typing import List, Dict, Tuple
# import pandas as pd
def remove_outliers(data: List[float], threshold: float = 3.0) -> List[float]:

    if len(data) < 2:
        return data

    mean = sum(data) / len(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    std_dev = math.sqrt(variance)


    if std_dev == 0:
        return data


    return [x for x in data if abs(x - mean) / std_dev <= threshold]
```

```python
# Activity 4
# from typing import List, Dict, Tuple
# Import pandas as pd
def train_test_split(data: List[float], ratio: float = 0.8) -> Tuple[List[float], List[float]]:

    if not 0 < ratio < 1:
        raise ValueError("Ratio must be between 0 and 1.")



    split_index = int(len(data) * ratio)

    train_set = data[:split_index]
    test_set = data[split_index:]

    return train_set, test_set
```
