```python
# from typing import List, Dict, Tuple
# import pandas as pd
data = [10, 20, 30]

def calculate_statistics(data:list[float])->dict[str, float]:
    mean = sum(data) / len(data)
    max_number = max(data)
    min_number = min(data)
    results = {"mean" : mean,
               "max" : max_number,
               "min" : min_number}
    if not data:
        return {"mean" : 0.0, "max" : 0.0, "min" : 0.0}
    return results
```
