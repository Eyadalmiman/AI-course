```python
# Activity 1
# from typing import List, Dict, Tuple
cat_a = [10, 20, 30]
cat_b = [40, 50, 60]

def calculate_stats(data:List[float])->Dict[str, float]:
    mean = sum(data) / len(data)
    max_number = max(data)
    min_number = min(data)
    results = {"mean" : mean,
               "max" : max_number,
               "min" : min_number}
    return results

stats_a = calculate_stats(cat_a)
stats_b = calculate_stats(cat_b)
print("A:", stats_a)
print("B:", stats_b)
```

```python
# Activity 2
# from typing import List, Dict, Tuple
# Import pandas as pd
data = [10, 20, 30]
def normalize_data(data:List[float], method:str=["minmax"])->List[float]:
    mean = sum(data) / len(data)
    std = sum((x - mean) ** 2 for x in data / len(data)) ** 0.5
    nor = (data - min(data)) / max(data) - min(data)
    zscore = (method - mean) / std
    return f"Normalized : {nor}, Standerized : {zscore}"

nor_stan = normalize_data(data, data)
print(nor_stan)
```
