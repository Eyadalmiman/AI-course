**Imports**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

**Code**

```python
feature = np.array([30.0, 50.0, 10.0])
weights = np.array([0.05, 0.8, -0.1])

fw = feature + weights
scale_fw = feature * weights
norm_f = np.linalg.norm(feature)
norm_w = np.linalg.norm(weights)

print(fw)
print(scale_fw)
print(norm_f)
print(norm_w)
print(f"feature shape : {feature.shape}")
print(f"weights shape : {weights.shape}")
```
