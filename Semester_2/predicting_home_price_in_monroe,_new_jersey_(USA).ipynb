```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv"
df = pd.read_csv(url)
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='blue',marker='+')
```

```python
new_df = df.drop('price',axis='columns')  # Input
new_df
```

```python
price = df.price     #output
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression() #create a model
reg.fit(new_df,price) # train the model
```

**(1) Predict price of a home with area = 3300 sqr ft**

```python
reg.predict([[3300]])
```

```python
reg.coef_     #m
```

```python
reg.intercept_   #b
```

**Y = m * X + b (m is coefficient and b is intercept)**

```python
3300*135.78767123 + 180616.43835616432
```

**(1) Predict price of a home with area = 5000 sqr ft**

```python
reg.predict([[5000]])
```

<h3 style="color:purple">Generate CSV file with list of home price predictions</h3>
