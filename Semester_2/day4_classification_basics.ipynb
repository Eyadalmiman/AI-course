## 1) Setup
Imports, a fixed seed, and a quick summary helper.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

RANDOM_STATE = 42
rng = np.random.default_rng(RANDOM_STATE)

def quick_summary(df, name):
    print(f"{name} shape: {df.shape}")
    print(df.head())
    print()

print("Setup complete ✅")
```

## 2) Tiny data / inputs for today's topics
We will create a small binary classification dataset with two features.

```python
n0, n1 = 70, 50
class0 = rng.normal(loc=[0, 0], scale=1.0, size=(n0, 2))
class1 = rng.normal(loc=[2.2, 2.0], scale=1.0, size=(n1, 2))

X = np.vstack([class0, class1])
y = np.array([0] * n0 + [1] * n1)

df = pd.DataFrame(X, columns=["feature_1", "feature_2"])
df["label"] = y

quick_summary(df, "df")
print("label counts:", df["label"].value_counts().to_dict())
```

## 3) Main task (guided)
Train a logistic regression classifier with a reproducible train/test split.

```python
# TODO 1: choose feature columns
feature_cols = ["feature_1", "feature_2"]

# TODO 2: set test_size (e.g., 0.2)
test_size = 0.2

# TODO 3: set random_state for reproducibility
random_state = RANDOM_STATE

X = df[feature_cols]
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=random_state, stratify=y
)

model = LogisticRegression()
model.fit(X_train, y_train)

print("coef:", model.coef_)
print("intercept:", model.intercept_)
```

## 4) Verify / evaluate
Check accuracy and a confusion matrix for sanity.

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", round(float(acc), 3))
print("Confusion matrix:")
print(cm)

print("Interpretation: accuracy shows how often the classifier is correct.")
```

## 5) Common pitfall + fix
Pitfall: treating classification like regression (using continuous predictions directly). Fix: threshold or use predict().

```python
proba = model.predict_proba(X_test)[:, 1]
wrong_preds = proba  # continuous probabilities
correct_preds = (proba >= 0.5).astype(int)

wrong_acc = accuracy_score(y_test, (wrong_preds > 0.5).astype(int))
correct_acc = accuracy_score(y_test, correct_preds)

print("Accuracy with thresholded probabilities:", round(float(correct_acc), 3))
print("Fix: convert probabilities to class labels before scoring.")
```
