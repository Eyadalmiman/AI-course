## Part 2 (5 min) — Supervised vs Unsupervised vs Reinforcement

Run the cell below. It creates:
- A **supervised** dataset (features + labels)
- An **unsupervised** dataset (features only)

```python
import numpy as np

rng = np.random.default_rng(7)

# Supervised: (X, y)
X_sup = rng.normal(size=(12, 2))
y_sup = (X_sup[:, 0] + 0.7 * X_sup[:, 1] > 0).astype(int)

# Unsupervised: X only
X_unsup = np.vstack([
    rng.normal(loc=(-2, -2), scale=0.7, size=(12, 2)),
    rng.normal(loc=( 2,  2), scale=0.7, size=(12, 2)),
])

# Reinforcement: tiny 1D walk to a goal
state = 0
goal = 3

def step(state, action):
    # action in {-1, +1}
    next_state = state + action
    reward = 1 if next_state == goal else -0.1
    done = next_state == goal
    return next_state, reward, done

X_sup.shape, y_sup.shape, X_unsup.shape, (state, goal)
```

## Part 3 (7 min) — The ML Workflow in Code: Data → Model → Predict → Evaluate

We’ll do a tiny **supervised classification** example.

**Tasks**
1) Train a model  
2) Predict on *training* and *unseen* data  
3) Compare **loss** and **accuracy** (why both matter)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss

# Data
X = X_sup
y = y_sup

# Split (quick demo for generalization)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=0, stratify=y
)

# Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict
proba_train = model.predict_proba(X_train)
proba_test  = model.predict_proba(X_test)
pred_train = model.predict(X_train)
pred_test  = model.predict(X_test)

# Evaluate
acc_train = accuracy_score(y_train, pred_train)
acc_test  = accuracy_score(y_test, pred_test)
loss_train = log_loss(y_train, proba_train)
loss_test  = log_loss(y_test, proba_test)

acc_train, acc_test, loss_train, loss_test
```

## Part 4 (3 min) — Loss & Optimization: One Tiny Gradient Descent Step (Optional)

Below is a **single-parameter** model \(\hat{y} = w x\) trained with **MSE**.

Run it, then change the learning rate `lr` and see what happens to the loss after one step.

```python
# Simple regression toy
x = np.array([1.0, 2.0, 3.0, 4.0])
y = np.array([2.0, 4.0, 6.0, 8.0])  # perfect y = 2x

w = 0.0
lr = 0.2

y_hat = w * x
loss_before = np.mean((y - y_hat)**2)

grad = (-2/len(x)) * np.sum(x * (y - y_hat))  # d/dw MSE
w_new = w - lr * grad

y_hat_new = w_new * x
loss_after = np.mean((y - y_hat_new)**2)

w, loss_before, w_new, loss_after
```
