```python
# Activity

# data
students = {
    "S001" : {
        "name" : "Alice Chen",
        "courses" : {
            "CS101" : {"grade" : 92, "credits" : 3},
            "MATH201" : {"grade" : 88, "credits" : 4},
            "AI301" : {"grade" : 95, "credits" : 3},
        },
        "advisor" : "Dr. Smith",
    },
    "S002" : {
        "name" : "Bob Lee",
        "courses" : {
            "CS101" : {"grade" : 85, "credits" : 3},
            "MATH201" : {"grade" : 90, "credits" : 4},
        },
        "advisor" : "Dr.Patel",
    }
}

coursesS001 = students["S001"]["courses"]
coursesS002 = students["S002"]["courses"]
B_GPA = sum(c["grade"] * c["credits"] for c in coursesS002.values()) // \
        sum(c["credits"] for c in coursesS002.values())
CS101_students = [x["name"] for x in students.values() if "CS101" in x["courses"]]
Alice_grades = sum(c["grade"] for c in students["S001"]["courses"].values())
Bob_grades = sum(c["grade"] for c in students["S002"]["courses"].values())
average_grades = (Alice_grades + Bob_grades) / 5
A_GPA = sum(c["grade"] * c["credits"] for c in coursesS001.values()) / \
        sum(c["credits"] for c in coursesS001.values())
best_name = None
best_gpa = -1.0
for s in students.values():
    courses = s["courses"]
    wsum = sum(c["grade"] * c["credits"] for c in courses.values())
    creds = sum(c["credits"] for c in courses.values())
    gpa = wsum / creds if creds else 0.0
    if gpa > best_gpa:
        best_gpa = gpa
        best_name = s["name"]
# prints
print(f"Alice Chen's AI301 grade : {students["S001"]["courses"]["AI301"]["grade"]}")
print(f"Bob Lee's GPA : {B_GPA}")
print(f"Students in CS101 : {CS101_students}")
print(f"The average grade across all courses is {average_grades}")
print(f"The person with the best gpa is {best_name} and their GPA is {best_gpa}")
```
