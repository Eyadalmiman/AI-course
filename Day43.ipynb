**Imports**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import simplejson
import random
from scipy import stats
from pathlib import Path
from typing import List, Dict, Tuple
```

**Code**

```python
mu = 100
sigma = 15

dist = stats.norm(loc = mu, scale = sigma)
a, b = 85, 115

prob_CDF = dist.cdf(b) - dist.cdf(a)
quantile_95 = dist.ppf(0.95)

sims = 1000000
samples = np.random.normal(mu, sigma, sims)
monte_carlo_prob = np.mean((samples >= a) & (samples <= b))

print(f"Theoretical probabiltity p({a} <= x <= {b}) : {prob_CDF:.5f}")
print(f"Monte carlo probability : {monte_carlo_prob:.5f}")
print(f"95th percent (cutoff) : {quantile_95:.2f}")
```
