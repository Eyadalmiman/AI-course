**Imports**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import simplejson
import random
from pathlib import Path
from typing import List, Dict, Tuple
```

**Code**

```python
trials = 100000000

coin_flip = np.random.choice(["Heads", "Tails"], size = trials)

target = "Heads"
count_target = np.sum(coin_flip == target)

emp_prob = count_target / trials
theory_prob = 0.5

print(f"The number of coin flips were {trials}")
print(f"The empirical probability is {emp_prob:.5f}")
print(f"The theoretical probability is {theory_prob:.5f}")
print(f"Difference : {abs(emp_prob - theory_prob):.5f}")
```
