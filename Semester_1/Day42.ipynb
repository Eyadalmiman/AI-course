**Imports**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import simplejson
import random
from pathlib import Path
from typing import List, Dict, Tuple
```

**Code**

```python
samples = 5000

uniform_data = np.random.uniform(0, 10, samples)
normal_data = np.random.normal(loc = 50, scale = 10, size = samples)

fig, axes = plt.subplots(1, 2, figsize = (12,5))

sns.histplot(uniform_data, kde = True, ax = axes[0])
axes[0].set_title("Uniform Distribution (0 to 10)")

sns.histplot(normal_data, kde = True, ax = axes[1])
axes[1].set_title("Normal Distribution (mean = 50, std = 10)")

plt.tight_layout()
plt.show()

print(f"Uniform - mean : {np.mean(uniform_data):.2f}, std : {np.std(uniform_data):.2f}")
print(f"Normal - mean: {np.mean(normal_data):.2f}, std : {np.std(normal_data):.2f}")
```
