**Code**

```python
df_norm = pd.DataFrame({
    "user_id" : [1, 2, 3, 4, 5, 6],
    "income_reported" : [55000, 62000, None, None, 48000, 0],
    "age" : [25, 30, None, 45, 28, None],
    "city" : ["Riyadh", None, "Jeddah", "Riyadh", None, "Dammam"]
})

df_imp = df_norm.copy()

df_imp["age_mean_imp"] = pd.to_numeric(
    df_imp["income_reported"], errors="ceorce"
)

df_imp["age_mean_imp"] = df_imp["age"].fillna(
    df_imp["age"].mean()
)

df_imp["age_median_imp"] = df_imp["age"].fillna(
    df_imp["age"].median()
)
```

```python
df = pd.DataFrame({
    "user_id" : [1, 2, 3, 4, 5],
    "age" : [25, np.nan, 30, np.nan, 28],
    "clicked_ad" : [1, 0, 1, 0, 1]
})

print("\n=== البيانات الأصلية ===\n")
print(df)

df["age_missing"] = df["age"].isna().astype(int)

df["age_imputed"] = df["age"].fillna(df["age"].median())

print("\n=== بعد إضافة Indicator وتعويض Median ===\n")
print(df)
```

```python
df = pd.DataFrame({
    "age" : [25, np.nan, 30, 45, np.nan, 28],
    "clicked_ad" : [1, 0, 1, 0, 1, 1]
})

age_mean = df["age"].mean()
X_mean = df["age"].fillna(age_mean).to_frame()
y = df["clicked_ad"].values

coef_mean = LinearRegression().fit(X_mean, y).coef_[0]
print(coef_mean)
```
